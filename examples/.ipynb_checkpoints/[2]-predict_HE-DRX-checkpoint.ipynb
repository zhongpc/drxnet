```python
from drxnet.funciton import load_pretrained_models, ensemble_prediction, voltageFeaturizer
from drxnet.drxnet.model import DRXNet
from drxnet.core import Featurizer
import numpy as np

import matplotlib.pyplot as plt
import matplotlib

font = {'family' : 'Arial',
        'weight' : 'regular',
        'size'   : 20}
matplotlib.rc('font', **font)

```

```python
model_list = load_pretrained_models()
vol_featurizer  = voltageFeaturizer()
```

## Plot multi-rate performance of Li$_{1.2}$Mn$_{0.1}$Mg$_{0.1}$Cr$_{0.30}$Ti$_{0.2}$Nb$_{0.1}$O$_{1.8}$F$_{0.2}$ (HE-1)

```python
plt.figure(figsize=(9,6))
composition_string  = 'Li1.2Mn0.1Mg0.1Cr0.3Ti0.2Nb0.1O1.8F0.2'

rate_list = [20, 100, 200, 400, 1000]
color_list = ['k', 'b', 'g', 'orange', 'red']

########################
Q_, V_ , *_ =  ensemble_prediction(model_list= model_list,
                                   vol_featurizer= vol_featurizer,
                                   composition_string= composition_string,
                                   input_rate= 10,
                                   input_cycle= 1,
                                   input_V_low = 1.5,
                                   input_V_high = 4.8)

plt.plot(Q_, V_, '--', color = 'k', linewidth = 2.0)

########################


for ii in range(len(rate_list)):

    Q_, V_,  *_ = ensemble_prediction(model_list= model_list,
                   vol_featurizer= vol_featurizer,
                   composition_string= composition_string,
                   input_rate= rate_list[ii],
                   input_cycle= 1,
                   input_V_low = 1.5,
                   input_V_high = 4.8)

    plt.plot(Q_, V_, color = color_list[ii], linewidth = 2.0)



plt.xlim(-10, 350)
plt.ylim(1.0, 5.0)

plt.xlabel("Capacity (mAh/g)", fontsize = 30)
plt.ylabel("Voltage (V)", fontsize = 30)

plt.xticks(np.arange(0, 351, 50))
plt.show()
```
