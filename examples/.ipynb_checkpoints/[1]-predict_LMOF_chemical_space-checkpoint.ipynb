```python
from drxnet.funciton import load_pretrained_models, ensemble_prediction, voltageFeaturizer
from drxnet.drxnet.model import DRXNet
from drxnet.core import Featurizer
import numpy as np

import matplotlib.pyplot as plt
import matplotlib

font = {'family' : 'Arial',
        'weight' : 'regular',
        'size'   : 20}
matplotlib.rc('font', **font)

```

```python
model_list = load_pretrained_models()
vol_featurizer  = voltageFeaturizer()
```

## Plot the 1st discharge capacity in Li$_{x}$Mn$_{2-x}$O$_{2-y}$F$_y$ space

```python
x_Li = np.linspace(1.17, 1.3333, 20)
y_F = np.linspace(0, 0.667, 15)
capacity_grid = np.zeros([20,15])

for ii, x in enumerate(x_Li):
    for jj, y in enumerate(y_F):
        comp_LMF = Composition.from_dict({'Li': x, 'Mn': 2- x, 'O':2-y, 'F':y})
        
        Q_, V_, *_ = ensemble_prediction(model_list= model_list,
                   vol_featurizer= vol_featurizer,
                   composition_string= comp_LMF, 
                   input_rate= 20,
                   input_cycle= 1,
                   input_V_low = 1.5,
                   input_V_high = 4.8)
        
        capacity_grid[ii, jj] = Q_[0]

```

```python

fig = plt.figure(figsize=(9,6))

plt.title(r'Li$_x$Mn$_{2-x}$O$_{2-y}$F$_y$ @ 1st')


plt.imshow(np.flipud(capacity_grid.T), cmap = 'plasma')


plt.xticks(np.array([0, 5, 10, 15, 20])-0.5, [1.17, 1.21, 1.25, 1.29, 1.33] )
plt.yticks(np.array([0, 3.75, 7.5, 11.25, 15])-0.5, [0.67, 0.5, 0.33, 0.17, 0])

cbar = plt.colorbar()
cbar.ax.get_yaxis().labelpad = 30

plt.show()
```

## Plot the 30th cycled discharge in Li$_{x}$Mn$_{2-x}$O$_{2-y}$F$_y$ space

```python
x_Li = np.linspace(1.17, 1.3333, 20)
y_F = np.linspace(0, 0.667, 15)
capacity_grid = np.zeros([20,15])

for ii, x in enumerate(x_Li):
    for jj, y in enumerate(y_F):
        comp_LMF = Composition.from_dict({'Li': x, 'Mn': 2- x, 'O':2-y, 'F':y})
        
        Q_, V_, *_ = ensemble_prediction(model_list= model_list,
                   vol_featurizer= vol_featurizer,
                   composition_string= comp_LMF, 
                   input_rate= 20,
                   input_cycle= 30,
                   input_V_low = 1.5,
                   input_V_high = 4.8)
        
        capacity_grid[ii, jj] = Q_[0]
```
